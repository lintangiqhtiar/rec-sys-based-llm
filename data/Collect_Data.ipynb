```python
from google_play_scraper import reviews, Sort
from datetime import datetime, timedelta

def fetch_recent_reviews(app_id, max_reviews=2000):
    """
    Fetch Google Play Store reviews in Indonesian from Indonesia for the last two years.

    :param app_id: str, the package name of the app (e.g., "com.example.app").
    :param max_reviews: int, max number of reviews to fetch.
    :return: list of dictionaries containing filtered review details.
    """
    all_reviews = []
    cutoff_date = datetime.now() - timedelta(days=730)  # Set cutoff to 2 years ago

    try:
        # Retrieve reviews in batches of 100
        for start in range(0, max_reviews, 100):
            result, _ = reviews(
                app_id,
                lang='id',        # Language: Indonesian
                country='id',     # Country: Indonesia
                sort=Sort.NEWEST,  # Sort by newest
                count=min(100, max_reviews - len(all_reviews))  # Get up to 100 at a time
            )

            # Filter reviews by date (last 2 years)
            for review in result:
                review_date = review['at']
                if review_date >= cutoff_date:
                    all_reviews.append(review)
                else:
                    return all_reviews  # Stop if we reach older reviews

            if len(all_reviews) >= max_reviews:
                break

        return all_reviews
    except Exception as e:
        print(f"An error occurred: {e}")
        return []
```

```python
# Usage
app_id = "id.qoin.korlantas.user"  # Replace with your app's package name
reviews_data = fetch_recent_reviews(app_id, max_reviews=2000)
for review in reviews_data:
    print(f"Date: {review['at']}, User: {review['userName']}, Rating: {review['score']}, Review: {review['content']}")
```

```python
import pandas as pd
```

```python
# Convert to DataFrame and save to CSV
if reviews_data:
    # Selecting only relevant fields for the CSV
    df = pd.DataFrame([{
        'date': review['at'],
        'user': review['userName'],
        'rating': review['score'],
        'review': review['content'],
        'thumbsUpCount': review['thumbsUpCount']
    } for review in reviews_data])

    # Save to CSV
    df.to_csv('korlantas_reviews.csv', index=False, encoding='utf-8')
    print("Reviews saved to korlantas_reviews.csv")
else:
    print("No reviews found.")
```
